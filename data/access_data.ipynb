## Egocentric-10K dataset

```python
# Define features
features = Features({
    'mp4': Value('binary'),
    'json': {
        'factory_id': Value('string'),
        'worker_id': Value('string'),
        'video_index': Value('int64'),
        'duration_sec': Value('float64'),
        'width': Value('int64'),
        'height': Value('int64'),
        'fps': Value('float64'),
        'size_bytes': Value('int64'),
        'codec': Value('string')
    },
    '__key__': Value('string'),    
    '__url__': Value('string')   
})

dataset = load_dataset(
    "builddotai/Egocentric-10K",
    data_files=[
        "factory_001/workers/worker_001/*.tar",
        "factory_001/workers/worker_002/*.tar"
    ],
    streaming=True,
    features=features
)
```

```python
import itertools

def get_video_by_index(dataset, index):
    """Get a specific video by index from streaming dataset"""
    return next(itertools.islice(iter(dataset["train"]), index, index + 1))
```

```python
def save_video(sample):
    print(sample["json"])
    factory_id = sample["json"]["factory_id"]
    worker_id = sample["json"]["worker_id"]
    video_index = sample["json"]["video_index"]
    video_filename = f"egocentric_video_{factory_id}_{worker_id}_{video_index}.mp4"

    # Save video to file
    print("Starting to save video")
    with open(video_filename, 'wb') as f:
        f.write(sample["mp4"])
    print("Video saved")
    Video(video_filename)
```

```python
VIDEO_INDEX = 5
sample = get_video_by_index(dataset, VIDEO_INDEX)
save_video(sample)
```
